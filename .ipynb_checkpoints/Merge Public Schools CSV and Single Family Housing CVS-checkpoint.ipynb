```python
# Dependencies
import pandas as pd
```

```python
# The path to our CSV file
public_schools = "data/Public_Schools.csv"
single_family_housing = "data/Zip_Zhvi_SingleFamilyResidence.csv"

# Read our public school data into pandas
ps_df = pd.read_csv(public_schools)
ps_df.head()

#read out single family housing data into pandas
fl_sfh_df = pd.read_csv(single_family_housing)
fl_sfh_df.head()
```

```python
#Pull out the Florida Public Schools
fl_ps_df = ps_df.loc[ps_df["STATE"] == "FL"]

print(len(fl_ps_df))
fl_ps_df.head()
```

```python
#reduce the number of columns
fl_ps_df = fl_ps_df.loc[:, ["NAME", "ADDRESS", "CITY",
                            "STATE", "ZIP", "WEBSITE", "LEVEL_", "ENROLLMENT", 
                            "ST_GRADE", "END_GRADE", "DISTRICTID", "FT_TEACHER", "SHELTER_ID"]]

fl_ps_df.head(10)
```

```python
# Remove all rows that aren't located in Florida
fl_sfh_df = fl_sfh_df.loc[fl_sfh_df["State"] == "FL", :]
fl_sfh_df.head(2)
```

```python
#reduce the number of columns
fl_sfh_df = fl_sfh_df.loc[:, ["RegionID", "RegionName", "City",
                            "State", "Metro", "CountyName", "SizeRank", "2019-01", 
                            "2019-02", "2019-03", "2019-04", "2019-05", "2019-06",
                             "2019-07", "2019-08", "2019-09", "2019-10", "2019-11", 
                            "2019-12"]]

fl_sfh_df.rename(columns={'RegionName':'ZIP'}, inplace=True)


fl_sfh_df.head(20)
```

```python
fl_sfh_df['tmp'] = 1
fl_ps_df['tmp'] = 1

merged_df = pd.merge(fl_sfh_df, fl_ps_df, on=['tmp'])
merged_df = merged_df.drop('tmp', axis=1)

merged_df.head()
```
